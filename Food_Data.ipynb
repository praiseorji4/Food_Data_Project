### Dealing with Different Quantities

```python
# Measurement Id value counts
print(df['quantity'].value_counts().index)
```

```python
# Function to convert quantity to kg
def convert_to_standard(quantity):
    if quantity in ['KG', 'L', 'Unit', 'Head', 'Gallon', 'Day', 'USD/LCU', 'Sack', 'Bunch', 'Loaf', 'Brush', 'Bar', 'Bundle', 'Packet', 'Pile', 'Heap', 'Package']:
        return float(1)
    elif 'KG' in quantity:
        return float(quantity.replace('KG', ''))
    elif 'G' in quantity:
        return float(quantity.replace('G', '')) / 1000
    elif 'ML' in quantity:
        return float(quantity.replace('ML', ''))/1000
    elif 'LCU/3.5kg' in quantity:
        return float(3.5)
    elif 'L' in quantity:
        return float(quantity.replace('L', ''))
    elif 'pcs' in quantity:
        return float(quantity.replace('pcs', ''))
    elif 'Tubers' in quantity:
        return float(quantity.replace('Tubers', ''))
    elif 'Pounds' in quantity:
        return float(quantity.replace('Pounds', ''))
    elif 'piece' in quantity:
        return float(quantity.replace('piece', ''))
    else:
        return None
# Apply the conversion function
df['quantity_in_standard'] = df['quantity'].apply(convert_to_standard)

# Calculate equivalent price per 1kg
df['equivalent_price_per_standard'] = df['price'] / df['quantity_in_standard']

df
```

#### Dealing with different currencies 

```python
# The issue with prices in multiple currencies is that we can't make one to one comparisons on the price column
# The solution? Create a new column for the price in a central currency (I will using the U.S Dollar as it is the easiest to get exch. rates on)
```

```python
# Exchange rates change over time so I will assume the prices are calculated at the end of each month
# Creating an end of month column
import calendar

# Creating a function that will take a year and a month, and supply the last date of that month
def last_date_of_month(year, month):
    # Get the last day of the given month
    _, last_day = calendar.monthrange(year, month)
    return pd.to_datetime(f"{year}-{month}-{last_day}")

# Creating a new column called 'LastDate' that will be the last date of the month
df['LastDate'] = df.apply(lambda row: last_date_of_month(row['year'], row['month']), axis=1)
```

```python
# They are quite a lot (to put it lightly). It will be only smart to categorize them

# Creating a function for categorization
def categorize_produce(produce):
    grains_and_cereals = ['Rice', 'Wheat', 'Maize', 'Millet', 'Sorghum']
    proteins = ['Meat', 'Eggs', 'Beans', 'Lentils', 'Fish']
    dairy = ['Milk', 'Cheese', 'Butter']
    fruits_and_vegetables = ['Apples', 'Bananas', 'Oranges', 'Onions', 'Tomatoes']
    oils_and_fats = ['Oil', 'Butter', 'Ghee']
    
    if any(item in produce for item in grains_and_cereals):
        return 'Grains and Cereals'
    elif any(item in produce for item in proteins):
        return 'Proteins'
    elif any(item in produce for item in dairy):
        return 'Dairy'
    elif any(item in produce for item in fruits_and_vegetables):
        return 'Fruits and Vegetables'
    elif any(item in produce for item in oils_and_fats):
        return 'Oils and Fats'
    else:
        return 'Others'

df['produce_category'] = df['produce'].apply(categorize_produce)
post_2010['produce_category'] = post_2010['produce'].apply(categorize_produce)
```

```python
correlations = {}
countries = post_2010['country'].unique()[:10]

for country in countries:
    subset = post_2010[post_2010['country'] == country]
    correlation = subset['price'].corr(subset['ExchangeRate'])
    correlations[country] = correlation

# print(correlations)
# Visualization: Bar chart showing correlations between exchange rates and produce prices
plt.figure(figsize=(10, 6))
plt.bar(correlations.keys(), correlations.values())
plt.xlabel('Country')
plt.ylabel('Correlation')
plt.title('Correlation Between Exchange Rates and Produce Prices by Country')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
